```python
import numpy as np
import imageio
from os import listdir
import albumentations as A
```

```python
def read_image(url):
    image = imageio.imread(url)
    return image

def augment_and_return(aug,image):
    aimg = aug(image=image)['image']
    return aimg

def save_augment_img(aimg,img,directory):
    uri = directory + '/' + img
    imageio.imwrite(uri,aimg)
```

```python
paper_input_path = '/content/drive/MyDrive/paper-scissors-rock/data/data/train/empty/'
paper_output_path = '/content/drive/MyDrive/paper-scissors-rock/data_augment1/empty/'
paper_input_images = listdir(paper_input_path)
paper_input_images.sort()
for i in paper_input_images:
  img_aug = A.Compose([
              A.HorizontalFlip(always_apply=False,p=0.5),
              A.ToGray(always_apply=False,p=1.0),
              A.Resize(always_apply=True, p=1.0, height=50, width=50, interpolation=0)
              ])
for i in paper_input_images:
    path2img = paper_input_path + i
    img = read_image(path2img)
    aimg = augment_and_return(img_aug,img)
    save_augment_img(aimg,i,paper_output_path)
```
